```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = "../data/PN chůze/"
Files = os.listdir(data_path)
```

```python
overview = pd.read_excel(data_path + '/PN chuze.xlsx')
```

```python
overview.rename(index=str, columns={"Unnamed: 0": "Filename", "Unnamed: 1": "Type"},inplace=True)
```

```python
overview.head()
```

```python
X = pd.DataFrame(columns=['PatientID','Type','N_parts','Length'])
for file in overview.Filename:
    #print(file)
    thisX = pd.DataFrame(columns=['PatientID','Type','N_parts','Length'])
    row = overview.loc[overview.Filename==file,:].dropna(axis=1)
    thisX.Type = row.Type
    thisX.N_parts = (row.size-2)/2
    thisX.PatientID = file[:-4]
    len = 0
    for n in range(2,row.size,2):
        len = len + row.iloc[0,n+1] - row.iloc[0,n]
    thisX.Length = len
    X = pd.concat([X,thisX],ignore_index=True)
```

### DATA CUT, based on excel table

```python
output_path = '../data/CUT/'
```

```python
for file in overview.Filename:
    print(file)
    thisdata = np.load(data_path+file[:-4] + '_filt.npy')
    row = overview.loc[overview.Filename==file,:].dropna(axis=1)
    for n in range(2,row.size,2):
        print('part ' + str(int((n-2)/2)))
        start = np.int(row.iloc[0,n])
        end = np.int(row.iloc[0,n+1])
        x = thisdata[:,start:end,:]
        np.save(output_path + file[:-4] + "_part" + str(int((n-2)/2)), x)
```
