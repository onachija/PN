```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = "../data/CALC/"
calcFiles = [f for f in os.listdir(data_path) if f.endswith('.calc')]
```

```python
calcFiles
```

```python
overview = pd.read_csv('../data/PN chuze/PN ataxie.csv')
overview.head()
```

```python
output_path = '../data/CALC_SEG/'
```

```python
for file in calcFiles:
    if file == 'N-10-1-Char00.calc':   # UPDATE ONLY THIS FILE, delete this IF to process all files
        print(file)
        thisdata = pd.read_csv('../data/CALC/' + file,header=2,sep='\t')
        thisdata.drop(columns='Unnamed: 946',inplace=True)
        row = overview.loc[overview.File==file[2:-5],:].dropna(axis=1)
        for n in range(5,row.size,2):
            print('part ' + str(int((n-2)/2)))
            start = np.int(row.iloc[0,n])
            end = np.int(row.iloc[0,n+1])
            x = thisdata.loc[start:end,:]
            x.to_csv(output_path + file[:-5] + "_part" + str(int((n-2)/2))+'.csv')
```
