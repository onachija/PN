```python
import numpy as np
import pandas as pd
import os
```

```python
data_path = "../data/CALC/"
calcFiles = [f for f in os.listdir(data_path) if f.endswith('.calc')]
```

```python
calcFiles
```

```python
overview = pd.read_excel('../data/PN chuze.xlsx')
```

```python
overview.rename(index=str, columns={"Unnamed: 0": "FilenameBVH", "Unnamed: 4": "Type"},inplace=True)
overview.head()
```

```python
for file in overview.FilenameBVH:
    overview.loc[overview.FilenameBVH == file,'File'] = file[:-4]
cols = overview.columns.tolist()
cols.insert(0, cols.pop(cols.index('File')))
overview = overview.reindex(columns= cols)
overview.drop(columns='FilenameBVH',inplace=True)
overview.head()
```

```python
output_path = '../data/CALC_SEG/'
```

```python
for file in calcFiles:
    print(file)
    thisdata = pd.read_csv('../data/CALC/' + file,header=2,sep='\t')
    thisdata.drop(columns='Unnamed: 946',inplace=True)
    row = overview.loc[overview.File==file[:-5],:].dropna(axis=1)
    for n in range(5,row.size,2):
        print('part ' + str(int((n-2)/2)))
        start = np.int(row.iloc[0,n])
        end = np.int(row.iloc[0,n+1])
        x = thisdata.loc[start:end,:]
        x.to_csv(output_path + file[:-5] + "_part" + str(int((n-2)/2))+'.csv')
```
