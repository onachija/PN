```python
import numpy as np
import pandas as pd
import os
from bvh import Bvh
```

```python
data_path = "../data/PN chuze/"
processed_path = "../data/PN chuze_processed/"
output_path = '../data/PARTS/'
Files = os.listdir(data_path)
```

```python
overview = pd.read_excel(data_path + '/PN chuze.xlsx')
overview.rename(index=str, columns={"Unnamed: 0": "Filename", "Unnamed: 1": "type"},inplace=True)
overview
```

```python
X = pd.read_csv(data_path + 'Overview.csv')
X
```

```python
#X = pd.DataFrame(columns=['subjectID','type','n_parts','total_samples'])
for file in overview.Filename:
    subject = file[:-4]
    if X['subjectID'].str.contains(subject).any():
        continue
    else:
        print(file)
        thisX = pd.DataFrame(columns=['subjectID','type','n_parts','total_samples'])
        row = overview.loc[overview.Filename==file,:].dropna(axis=1)
        thisX.type = row.type
        thisX.n_parts = (row.size-2)/2
        thisX.subjectID = file[:-4]
        len = 0
        for n in range(2,row.size,2):
            len = len + row.iloc[0,n+1] - row.iloc[0,n]
        thisX.total_samples = len
        X = pd.concat([X,thisX],ignore_index=True)
```

```python
X
```

```python
for file in Files:
    if file.endswith('.bvh'):
        if np.isnan(X.loc[X.subjectID==file[:-4],'frame_time'].values[0]):
            print(file)   
            with open(data_path + file) as f:
                mocap = Bvh(f.read())
                print(mocap.frame_time)
                X.loc[X.subjectID==file[:-4],'frame_time'] = mocap.frame_time
```

```python
X
```

```python
X.to_csv(data_path + 'Overview.csv',index=False)
```

```python
X = pd.read_excel('../data/PN chuze/PN chuze.xlsx')
X.rename(index=str, columns={"Unnamed: 0": "FilenameBVH", "Unnamed: 1": "Type"},inplace=True)
overview = pd.read_csv('../data/PN chuze/Overview.csv')
```

```python
for file in X.FilenameBVH:
    subject = file[:-4]
    X.loc[X.FilenameBVH == file,'File'] = str.upper(X.loc[X.FilenameBVH == file,'Type'][0][0]) + '-' + subject.replace('_','-')
    X.loc[X.FilenameBVH == file,'Ts [s]'] = overview.loc[overview.subjectID == subject,'frame_time'].values
    X.loc[X.FilenameBVH == file,'fs [Hz]'] = 1/(overview.loc[overview.subjectID == subject,'frame_time'].values)
    X.loc[X.FilenameBVH == file,'Frames'] = overview.loc[overview.subjectID == subject,'total_samples'].values
cols = X.columns.tolist()
cols.insert(0, cols.pop(cols.index('File')))
cols.insert(1, cols.pop(cols.index('Ts [s]')))
cols.insert(2, cols.pop(cols.index('fs [Hz]')))
cols.insert(3, cols.pop(cols.index('Frames')))
X = X.reindex(columns= cols)
X.drop(columns='FilenameBVH',inplace=True)
X
```

```python
X.to_csv('../data/PN chuze/PN ataxie.csv',index=False)
```
