```python
with open('../data/PN chůze/681001_Char00.bvh') as f:
    mocap = Bvh(f.read())
```

```python
next(mocap.root.filter('ROOT'))['OFFSET']
```

```python
number_frames = mocap.nframes
number_frames
```

```python
mocap.frame_time
```

```python
[str(item) for item in mocap.root]
```

```python
joint_names = mocap.get_joints_names()
joint_names
```

```python
#Get single JOINT name
joint_data = np.zeros((len(joint_names),number_frames,6))
joint_data_filt = np.zeros((len(joint_names),number_frames,6))
i=0
for joint_name in joint_names:
    print(joint_name)
    joint_data[i] = mocap.frames_joint_channels(joint_name, ['Xposition', 'Yposition', 'Zposition', 'Yrotation', 'Xrotation', 'Zrotation'])
    i=i+1
```

```python
from scipy.signal import savgol_filter
for i in range(len(joint_names)):
    sg = savgol_filter(joint_data[i].T, 1001, 2)
    joint_data_filt[i] = sg.T
    
np.save("681001_raw", joint_data)
np.save("681001_filt", joint_data_filt)
```

```python
plt.plot(joint_data[0,5000:15000,1])
plt.plot(joint_data_filt[0,5000:15000,1],"r")
plt.xlabel("Time (millisecond)")
plt.ylabel("Acceleration (m/s" + r'$^2$)')
```
